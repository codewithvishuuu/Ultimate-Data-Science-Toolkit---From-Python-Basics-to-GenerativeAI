```python
import numpy as np

arr_1D = np.array([1,2,3,4,5])
print("1D array : ",arr_1D)

arr_2D = np.array([[1,2,3,6,5],
  [6,7,8,9,10]])
print("2D array : ",arr_2D)
```

```python
py_list = [1,2,3]
print("Python list multiplication :",py_list * 2)

np_array = np.array([1,2,3]) #ELEMENTS WISE multiplication
print("Python list multiplication :",np_array * 2)

import time
start = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n List operation time :",time.time() - start)

start = time.time()
np_array = np.arange(1000000) * 2
print("\n Numpy operation time :",time.time() - start)

```

```python

zeros = np.zeros((3,4))
print("Zeros array : \n",zeros)

ones = np.ones((3,3))
print("ones array : \n",ones)

full = np.full((2,3), 7)
print("full array : \n",full)

random = np.random.random((2,3))
print("random array : \n",random)

sequence_array = np.arange(0, 10, 2)
print("sequence array : \n",sequence_array)
```

```python
vactor = np.array([1,2,3,4,5])
print("Vector : ",vactor)

matrix = np.array([[1,2,3],
                   [4,5,6]])
print("Matrix : ",matrix)

tensor = np.array([[[1,2], [3,4]],
                  [[5,6], [7,8]]])
print("Tensor : ",tensor)
```

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

#Properties :---
print("Shape :", arr.shape)
print("Dimension :", arr.ndim)
print("Size :", arr.size)
print("DType :", arr.dtype)
```

```python

arr =np.arange(12)
print("Original array :", arr)

reshaped = arr.reshape((3,4))
print(" \n Reshaped array :", reshaped)

flattened = reshaped.flatten() # retuns copy
print(" \n Flat array :", flattened)

# revel (returns view, instead of copy)
raveled = reshaped.ravel()
print(" \n reveled array :", raveled)

# Transpose
transpose = reshaped.T
print(" \n Transposed array :", transpose)
```
